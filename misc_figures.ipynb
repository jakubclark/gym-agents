```python
from matplotlib import pyplot as plt
from math import cos, pi
from numpy import arange
```

```python
positions = [i for i in arange(-1.2, 0.6, 0.01)]
rewards = [round(1 * (position + 1.2) if position > -0.2 else -1., 3) for position in positions]
track = [cos(3 * pos - (pi / 2)) for pos in positions]
```

```python
plt.plot(positions, rewards)
plt.xlabel('Position')
plt.ylabel('Reward')
plt.title('Cart Position vs Reward')
plt.legend(('Reward', 'Track'))
plt.show()
```

```python
plt.plot(positions, rewards)
plt.plot(positions, track)
plt.xlabel('Position')
plt.ylabel('Reward')
plt.title('Cart Position vs Reward')
plt.legend(('Reward', 'Track'))
plt.show()
```
